```python
df[colu] = df[colu].astype(str)
```

```python
df.index = df.Id
```

```python
df.drop(['Id','Medical_History_10','Medical_History_32','Medical_History_24',
'Medical_History_15',
'Family_Hist_5',
'Family_Hist_3',
'Family_Hist_2',],axis=1,inplace=True)
```

```python
df.fillna(df.mean, axis=1, inplace=True)
```

```python
dummy = pd.get_dummies(df[colu], drop_first =True)
```

```python
numeric = df.select_dtypes(exclude=['object'])
```

```python
train = numeric.merge(dummy, left_index=True, right_index=True)
```

```python
train.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
y = train.loc[:,['Response']]
```

```python
train.drop('Response', axis=1, inplace=True)
```

```python
train.shape
```

```python
X, y  = train.values, y.values
```

```python

X_train , X_val, y_train, y_test = train_test_split(X, y, test_size=.1, random_state=23)
```

```python
pca = PCA(n_components=.99,)
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
```

```python
X_train_pca.shape
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(class_weight='balanced', penalty='l2',verbose=2)
```

```python
lr.fit(X_train, y_train.ravel())
```

```python
lr.score(X_val, y_test)
```

```python
lr.score(X_train, y_train)
```

```python
y_train.ravel()
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC()
```

```python
sv.fit(X_train_pca, y_train.ravel())
```

```python
sv.score(X_val_pca, y_test)
```

```python
sv.score(X_train_pca, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_range = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0 , 100.0 ,1000.0]
param_grid = [{'svc__C':param_range, 'svc__kernel':['linear']},
               {'svc__C':param_range, 'svc__kernel':['rbf'], 'svc__gamma':param_range}]

gs = GridSearchCV(estimator=sv,
                 param_grid=param_grid,
                 scoring='accuracy',
                 cv=10)

gs = gs.fit(X_train_pca, y_train.ravel())
print('Best average Precision : ', gs.best_score_)
print(gs.best_params_)
```
