- SVM can be __kernelized__ to solve non linear classification problems.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
```

First, we create the data to depict the nonlinear problem

```python
np.random.seed(1)
X_xor = np.random.randn(200,2)
```

```python
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                      X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)
```

```python
plt.scatter(X_xor[y_xor == 1, 0],
           X_xor[y_xor == 1, 1],
           c = 'b', marker='x', label='1')

plt.scatter(X_xor[y_xor == -1, 0],
           X_xor[y_xor == -1, 1],
           c = 'r', marker='s', label='-1')

plt.xlim([-3, 3])
plt.ylim([-3, 3])
plt.legend(loc = 'best')
plt.show()
```

_Obviously_ we would not be able to separate samples of positive and negative classes very well using a _linear hyerperPlane_

- The idea behind __kernle methods__ to deal with such linearly inseprabale data is to __create nonlinear combination of the original features to project them into a higher-dimensional space via mapping function $\phi$__

<center>$\phi \left (x_1, x_2 \right ) \space = \space \left( z_1, z_2,z_3 \right) = \left (x_1, x_2, x_1^2 + x_2^2 \right) $ 

One problem, with this mapping approach is that the construction of the new feature is _computationally very exprensive_, especially __if we are dealing with high-dimensional data__ 
- This is where the __kernel trick__ comes into play, in practice all we need to replace the dot product<br> <center>$x^{(i)T} x^{(j)} $ by $\phi \left (x^{(i)} \right )^T \phi \left (x^{(j)}\right)$</center>

- In order to save the expensive step of calculating this dot product between two points explictily, we defind a so-called  __kernel function :__
    <center>K $\left ( x^{(i)}, x^{(j)} \right) = \phi\left(x^{(i)}\right)^T\phi \left(x^{(j)}\right)$</center>

- one of the widely used kernel is __Radial Basis Function (RBF)__ kernel or simply called the __Gaussian Kernel.__

<center>K $\left ( x^{(i)}, x^{(j)} \right) = exp \left ( - \frac {|| x^{(i)} \space - \space x^{(j)} \space ||^2}{2 \sigma^2}\right)$</center>
<br>This is often simplified to
<center>K $\left ( x^{(i)}, x^{(j)} \right) = exp \left (- \space \gamma \space {|| x^{(i)} \space - \space x^{(j)} \space ||^2}\right)$</center>
Here, $\gamma \space = \frac {1} {2\sigma^2}$ is free parameter _to be optimised_

__Note:__ Roughly speaking , the term kernel can be interpreted as a __similarity function__ between a pair of samples.
- The minus sign inverts the __distance measure into a similarity score__, and, due to the exponential term , the resulting similarity score will fall into a range between 1 (for exactly similar samples) and 0 (for very dissimilar samples)

- The $\gamma $ parameter, which we set to gamma=0.1, can be understood as cut-off  parameter for the Gaussian sphere.
- If we __increase values for $\gamma $, __ we __increase the influence or reach of the training samples__, which leads to a tighter and bumpier decision boundary

##### Let's apply the SVM with RBF kernel on iris flower dataset to see the impact of $\gamma $ parameter

```python
X_train_std, X_test_std, y_train, y_test = get_iris(standarized=True)

X_combined_std = np.vstack((X_train_std,X_test_std))
y_combined = np.hstack((y_train,y_test))
```

```python
svm = SVC(kernel='rbf', random_state=1, gamma=0.10, C=10.0)
svm.fit(X_train_std,y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105,150))
plt.xlabel('petal length [standarized]')
plt.ylabel('petal width [standarized]')
plt.legend(loc = 'lower right')
plt.show()
```

Since, we choose a small value for $\gamma$, the resulting decision boundary of the RBF kernel SVM model is relatively soft.

```python
svm = SVC(kernel='rbf', random_state=1, gamma=10, C=10.0)
svm.fit(X_train_std,y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105,150))
plt.xlabel('petal length [standarized]')
plt.ylabel('petal width [standarized]')
plt.legend(loc = 'lower right')
plt.show()
```

```python
svm = SVC(kernel='rbf', random_state=1, gamma=100, C=10.0)
svm.fit(X_train_std,y_train)
```

```python
plot_decision_regions(X_combined_std, y_combined, classifier=svm, test_idx=range(105,150))
plt.xlabel('petal length [standarized]')
plt.ylabel('petal width [standarized]')
plt.legend(loc = 'lower right')
plt.show()
```

As we increase the value of $\gamma$, we observe the decision boundary around the classes 0 and 1 becoming much tighter using a relatively large value of $\gamma$

- __Note: __ This illustrates that the ___$\gamma$ parameter, also play an important role in controlling overfitting.___
